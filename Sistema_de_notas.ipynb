```python

# Sistema de Gestão de Notas

# Função para cadastrar notas de um aluno
def cadastrar_notas():
    notas = []
    while True:
        try:
            nota = float(input("Digite a nota do aluno (ou -1 para encerrar): "))

            # Condição de parada
            if nota == -1:
                break

            # Garantindo que a nota seja válida
            if 0 <= nota <= 10:
                notas.append(nota)
            else:
                print(" A nota deve estar entre 0 e 10.")
        except ValueError:
            print(" Digite um valor numérico válido.")

    return notas


# Função para calcular a média
def calcular_media(notas):
    if len(notas) == 0:
        return 0
    return sum(notas) / len(notas)


# Função para determinar a situação
def verificar_situacao(media):
    return "Aprovado " if media >= 7 else "Reprovado "


# Função para exibir o relatório final de 1 aluno
def relatorio_individual(nome, notas, media, situacao):
    print("\n===== RELATÓRIO DO ALUNO =====")
    print(f"Aluno: {nome}")
    print(f"Notas: {notas}")
    print(f"Média: {media:.2f}")
    print(f"Situação: {situacao}")
    print("===============================")


# Programa principal
def main():
    alunos = []  # Lista para armazenar os dados de todos os alunos

    print(" Sistema de Gestão de Notas de Alunos ")

    while True:
        nome = input("\nDigite o nome do aluno (ou 'sair' para encerrar): ")

        if nome.lower() == "sair":
            break

        # Cadastro das notas
        notas = cadastrar_notas()

        # Cálculo da média
        media = calcular_media(notas)

        # Situação
        situacao = verificar_situacao(media)

        # Guardar dados do aluno
        alunos.append({
            "nome": nome,
            "notas": notas,
            "media": media,
            "situacao": situacao
        })

        # Relatório individual
        relatorio_individual(nome, notas, media, situacao)

    # Resumo final da turma
    print("\n===== RESUMO FINAL DA TURMA =====")
    for aluno in alunos:
        print(f"{aluno['nome']} - Média: {aluno['media']:.2f} - {aluno['situacao']}")
    print("=================================")


# Executa o programa
main()
```
